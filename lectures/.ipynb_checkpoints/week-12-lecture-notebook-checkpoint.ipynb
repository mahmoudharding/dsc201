## User-Defined Functions

In Python, a user-defined function is a reusable block of code created with the `def` keyword to perform a specific task. Functions make code more organized and efficient by allowing you to define logic once and reuse it. They can take inputs, process data, and return outputs.

### The Parts of a User-Defined Function 

In R, a user-defined function is a reusable block of code created with the `function()` keyword to perform a specific task. Functions help make your code cleaner and more efficient because you can define logic once and reuse it multiple times. They can take input values, perform operations, and return an output.

### The Parts of a User-Defined Function

An R function has several key components:

* Function definition (`function()`): The keyword that tells R you are defining a new function.
* Function name: A clear, descriptive name that tells what the function does (e.g., `calculate_average`).
* Parameters: Variables inside parentheses that receive input values when the function is called.
* Curly braces (`{ }`): Indicate the start and end of the function body.
* Function body: The code inside `{ }` that runs when the function is called.
* Return value (`return()` or last evaluated expression): Sends a result back to where the function was called. In R, you can explicitly use `return()`, or simply rely on the last line being returned automatically.

For example:

```r
calculate_average <- function(numbers) {
  # Calculates the average of a numeric vector.
  #
  # Parameter:
  #   numbers: A numeric vector.
  #
  # Returns:
  #   The average of the numbers.

  total <- sum(numbers)
  average <- total / length(numbers)
  return(average)
}
```

The functions below can be used to create new features, such as predicting the crab’s likely species or identifying whether it was farm-raised or caught in the wild.

````r
classify_species <- function(d) {
  # Classify the crab species based on its diameter.
    
  # Parameter:
  #   d (numeric): Diameter of the crab in feet.
  
  # Returns:
  #   A character string indicating the likely species of the crab.

  case_when(
    d < 0.3 ~ "Juvenile",
    d < 0.75 ~ "Blue",
    d < 0.9 ~ "Dungeness",
    d < 1.3 ~ "Red King",
    TRUE ~ "Tasmanian"
  )
}


classify_farming_status <- function(d) {
  # Classify whether a crab is likely farm-raised or caught in the wild 
  # based on its diameter.
  
  # Parameter:
  #   d(numeric): Diameter of the crab in feet.
  
  # Returns:
  #   1(int) if likely farm-raised, 0(int) if likely caught in the wild.

  if_else(
    d >= 0.4 & d <= 0.9,
    1,
    0
  )
}
````

````r
crabs <- crabs %>%
  mutate(
    species = classify_species(diameter),
    farming_status = classify_farming_status(diameter)
  )
````

Verify that the new features were added to the `crabs` dataset. 

````r
head(crabs)
````

## Purpose and Fundamentals of Data Visualization

Data visualization is the practice of representing data graphically to reveal patterns, trends, and insights that might be difficult to interpret from raw numbers alone. Effective visualizations simplify complex information, making it easier to communicate findings, support decision-making, and identify relationships within data. The fundamentals of data visualization include selecting appropriate graphical representations, ensuring clarity and accuracy, and emphasizing key insights while avoiding misleading representations (Tufte 2001). 

While data visualization involves many concepts, techniques, and considerations, our discussion will focus on fundamental visualization types commonly used in exploratory data analysis, including bar charts, histograms, box plots, line charts, and scatter plots. These visualizations can be created in both Python and R. 

### `ggplot2`

`ggplot2` is a data visualization package in R that follows the “grammar of graphics,” a layered approach to building plots. Instead of creating a complete graph with a single command, you construct it step by step. First specifying the data and aesthetic mappings, then adding layers like points, lines, labels, and themes.

```
library(ggplot2)
```

A typical `ggplot2` visualization begins with the `ggplot()` function, which defines the **data** and **aesthetics** (such as which variable goes on the $x-$ or $y-$axis). You then add layers using functions like `geom_point()` for scatterplots or `geom_bar()` for bar charts. Additional functions like `labs()` and `theme()` let you adjust titles, axis labels, and visual style for further customization.

**Note:** To learn more about visualizations in `ggplot2`, click [here](https://ggplot2.tidyverse.org/).

Import the necessary libraries for creating our visualizations.

````r
library(ggplot2)
````

## Bar Chart

A bar chart is a graphical representation of categorical data where each category is represented by a bar, with the height or length of the bar corresponding to its value. It is used to compare the frequency, count, or other metrics across different categories. Bar charts are ideal when you need to visually compare discrete categories or show trends over time.

**Example 1.** Create a bar chart to display the frequency of each category.

````r
g <- ggplot(data = crabs, aes(x = sex))
g + geom_bar()
````

**Example 2.** Add a title to your bar chart.

````r
g + 
  geom_bar() +
  labs(title = "Distribution of Sex", x = "Sex", y = "Frequency")
````

````r
g + 
  geom_bar() +
  labs(title = "Distribtuion of Sex", x = "", y = "Frequency") +
  scale_x_discrete(labels = c(
      "F" = "Female",
      "I" = "Indeterminate",
      "M" = "Male"
  )
                  )
````

**Example 3.** Create a stacked bar chart.

````r
g <- ggplot(data = crabs, aes(x = species, fill = sex))
  
g + geom_bar()
````

**Example 4.** Customize the legend by adding a title and labels, then include clear axis labels.

````r
g +
  geom_bar() +
  scale_fill_discrete(name = "Crab Sex",
                      labels = c("Female", "Indeterminate", "Male")
                     )
````

Unlike a bar chart, which displays categorical data, a histogram groups continuous data into bins (ranges) and shows the frequency or count of data points within each bin. The height of each bar represents the number of data points that fall within each bin.

## Box Plot

A boxplot is a way of displaying the distribution of data based on a five-number summary. 

- Median (Q2): The line inside the box represents the median (the middle value of the data set).

- First quartile (Q1): The lower edge of the box, representing the 25th percentile (where 25% of the data lies below this value).

- Third quartile (Q3): The upper edge of the box, representing the 75th percentile (where 75% of the data lies below this value).

- Interquartile Range (IQR): The range between the first quartile (Q1) and third quartile (Q3), which contains the middle 50% of the data.

- Whiskers: These extend from the edges of the box to the smallest and largest values within 1.5 times the IQR from Q1 and Q3, respectively. They represent the range of most of the data.

- Outliers: Data points outside the whiskers are considered outliers and are usually plotted as individual dots.

### Factors

In R, a factor is a special data type used to represent categorical variables, such as groups, labels, or categories (e.g., "Male/Female" or "Small/Medium/Large"). 

Factors tell R to interpret values as discrete categories rather than as continuous numbers. In a boxplot, the grouping variable must be categorical, regardless of whether it is placed on the $x-$axis or $y-$axis, while the numeric variable goes on the opposite axis to show the distribution. If a category variable is stored as numbers (such as 1 or 0 for farming status), R may mistakenly treat it as continuous. Converting it using `as.factor()` ensures that `ggplot2` recognizes it as a true categorical variable. The resulting visualization contains boxplots for each group rather than plotting the values across a numeric scale.

**Note:** `as.factor()` is one of several `as.` functions in R used to convert a variable’s data type. Others include `as.numeric()`, `as.character()`, and `as.logical()`.

**Example 7.** Create a box plot to display the distribution and show the median, quartiles, and any potential outliers.

````r
g <- ggplot(data = crabs, aes(x = diameter)) 

g + geom_boxplot()
````

**Example 8.** Create a box plot that compares crab shucked weight across sex categories.

````r
g <- ggplot(data = crabs, aes(x = diameter, y = factor(sex)))

g + geom_boxplot()
````

**Example 9.** Customize your boxplot by adding a title and axis labels and replacing the default $x-$axis labels with full names.

````r
g +
  geom_boxplot(color = 'black', fill = 'lightblue', linewidth = 0.125) + 
  labs(title = "Distribution of Diameter by Sex", 
       x = "",
       y = "") +
  scale_y_discrete(labels = c(
      "F" = "Female",
      "I" = "Indeterminate",
      "M" = "Male")
                  )
````

## Line Plot

A line plot is used to show trends or changes over time by connecting data points with straight lines. It’s especially useful when visualizing how a variable changes across a continuous interval such as days, months, or years. Before creating a line plot, it’s common to summarize or aggregate your data — for example, by computing the average or total for each time period.

In R with `dplyr` + `ggplot2`, this is done using `group_by()` and `summarize()`:


### Flights

Our dataset, called the "Flights" dataset, belongs to the Seaborn library. This dataset provides a monthly tally of airline passengers from 1949 to 1960.

The Flights dataset comprises three distinct columns:

- year: Represents the year in which the count of passengers was taken.
- month: Points towards the month in which the passenger count was gathered.
- passengers: Indicates the number of passengers that traveled in that month of a particular year.

````r
flights = read.csv('data/flights.csv')
head(flights)
````

**Example 15.** Create a line plot to visualize how the maximum height of skyscrapers has evolved over time.

````r
g <- ggplot(flights, aes(x = year, y = passengers, group = 1)) 

g +
  geom_line() +
  labs(
    title = "Passengers by Year",
    x = "Year",
    y = "Number of Passengers"
  )
````

Since there are multiple observations for the same year, we need to group the data by `year` and choose a summary statistic to compute for each group.

When grouping data by year (or any other category), you can choose from several common statistical measures depending on your goal.

* **Mean**

  ```r
  flights %>%
    group_by(year) %>%
    summarise(mean_passengers = mean(passengers, na.rm = TRUE))
  ```


* **Median**

  ```r
  flights %>%
    group_by(year) %>%
    summarise(median_passengers = median(passengers, na.rm = TRUE))
  ```


* **Sum**

  ```r
  flights %>%
    group_by(year) %>%
    summarise(total_passengers = sum(passengers, na.rm = TRUE))
  ```


* **Count (rows per year)**

  ```r
  flights %>%
    group_by(year) %>%
    summarise(n_rows = n())
  ```


* **Min / Max**

  ```r
  flights %>%
    group_by(year) %>%
    summarise(
      min_passengers = min(passengers, na.rm = TRUE),
      max_passengers = max(passengers, na.rm = TRUE)
    )
  ```

The statistic you choose depends on what you’re analyzing: mean/median for central tendency, sum for totals, count for volume of records, and min/max for extremes or ranges,

````r
dat <- flights %>%
  group_by(year) %>%
  summarise(total_passengers = sum(passengers, na.rm = TRUE))

g <- ggplot(dat, aes(x = year, y = total_passengers)) 

g +
  geom_line() +
  labs(
    title = "Total Number of Passengers per Year",
    x = "Year",
    y = "Passengers"
  )
````
