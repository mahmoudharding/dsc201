### The Hypotheses

We will use a hypothesis test to determine whether the difference in birthweights between babies of smokers and non-smokers is real or just due to chance.

**Just due to chance:** Babies of smokers and non-smokers have the same birthweights, and any difference in the sample is random.

**Not due to chance:** Babies of smokers have lower birthweights than babies of non-smokers, and the difference is real.

### The Test Statistic

The **Not due to chance** hypothesis says that the average birthweight of babies born to mothers who smoke is smaller than the average birthweight of babies born to mothers who do not smoke. Because we are comparing the two averages, it makes sense to use the difference between the group means as our statistic.

```python
average birthweight of smokers  – average birthweight of non-smokers
```
